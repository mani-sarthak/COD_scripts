```python
import json
json_file = '*.json' ### add the json mapping here !!

with open(json_file, "r") as json_file:
    data = json.load(json_file)
    
data = data['features']
```

```python
d = dict()
for x in data:
    state = x['properties']['state']
    station_name = x['properties']['station_name']
    wscode = x['properties']['wscode']
    coordinates = x['geometry']['coordinates']
    ########## this 4 is for fine tuning 
    ####### DO NOT CHANGE !!
    d[(state, station_name,  round( coordinates[1], 4), round(coordinates[0], 4),)] =  wscode
    
print(d)
print(df)
```

```python
import csv
input_file = '*.csv' ### Path to csv file for which mapping is to be done !!
output_file = 'out.csv'

analysis = dict()

with open(input_file, 'r') as csv_in, open(output_file, 'w', newline='') as csv_out:
    reader = csv.reader(csv_in)
    writer = csv.writer(csv_out)

    header = next(reader)
    header.append('watershed_code')
    writer.writerow(header)

    for row in reader:
        new_value = d.get((row[3], row[2], round(float(row[0]), 4), round(float(row[1]), 4)), None)
        analysis[new_value] = analysis.get(new_value, 0) + 1
        row.append(new_value)
        writer.writerow(row)
```

```python
mp = [(analysis[v], v) for v in analysis]


mp.sort(reverse=True)

print(mp)    
```
